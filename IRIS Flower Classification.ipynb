```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv.csv")
df.head()
```

```python
df.tail()
```

```python
df.columns
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df.SepalLengthCm,df.SepalWidthCm,alpha=0.9,color='Red')
plt.title('Iris Sepal Dimensions')
plt.xlabel('SepalLengthC')
plt.ylabel('SepalWidthCm')
plt.grid(True ,linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['SepalLengthCm'], alpha=0.7, color='purple', bins=15, label='SepalLengthCm')
plt.hist(df['SepalWidthCm'], alpha=0.7, color='red', bins=15, label='SepalWidthCm')
plt.title('Distribution of Sepal Dimensions')
plt.xlabel('Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(df['SepalLengthCm'], linestyle='dashed', marker='o', alpha=0.6, label='Sepal Length', color='red')
plt.xlabel('Measurement Index')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='dotted', alpha=0.5)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6,6))
fig.suptitle('Comparison of Sepal and Petal Lengths', fontsize=14)
axes[0].plot(df['SepalLengthCm'], color='blue', label='Sepal Length')
axes[0].set_title('Sepal Length Distribution')
axes[0].set_xlabel('Sepal')
axes[0].set_ylabel('Length (cm)')
axes[0].legend()

axes[1].plot(df['PetalLengthCm'], color='orange', label='Petal Length')
axes[1].set_title('Petal Length Distribution')
axes[1].set_xlabel('Petal')
axes[1].set_ylabel('Length (cm)')
axes[1].legend()

fig.tight_layout()
fig.subplots_adjust(top=0.85)
plt.show()
```

```python
species_counts = df['Species'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(species_counts, labels=species_counts.index, colors=['skyblue', 'lightcoral', 'gold'], autopct='%1.1f%%', startangle=140)
plt.title('Iris Species Distribution')
plt.show()
```
